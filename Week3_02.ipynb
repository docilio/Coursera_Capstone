```python
import pandas as pd
```

```python
import requests
import io

url = "https://raw.githubusercontent.com/docilio/Coursera_Capstone/master/Geospatial_Coordinates.csv"
s = requests.get(url).content
ds = pd.read_csv(io.StringIO(s.decode('utf-8')))


url2 = "https://raw.githubusercontent.com/docilio/Coursera_Capstone/master/Toronto.csv"
s2 = requests.get(url2).content
ds2 = pd.read_csv(io.StringIO(s2.decode('utf-8')))
```

```python
ds.set_index('Postal Code')
ds.head()
```

```python
ds2.set_index('Postcode')
ds2.head()
```

```python
Combined = ds2.join(ds)
Combined.head()
```

```python
#extra work:
import base64
import pandas as pd
from IPython.display import HTML

def create_download_link( df, title = "Download CSV file (Combined)", filename = "data.csv"):
    csv = df.to_csv()
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

a = create_download_link(Combined,filename='Combined.csv')

a
```
